```python
# Import packages
from uszipcode import ZipcodeSearchEngine
import pandas as pd
import numpy as np
```

```python
market = pd.read_csv("data/farmersmarkets-or.csv")
market.head()
```

```python
market.iloc[1]["zip"]
```

```python
zip_null_idx = np.where(pd.isnull(market["zip"]))
zip_null_idx[0][:10]
print(len(*zip_null_idx))
```

```python
search = ZipcodeSearchEngine()
for i in list(*zip_null_idx):
    print(i)
    temp = market.iloc[i]
    x= temp["x"]
    y= temp["y"]
    if(pd.notna([x,y]).all()):
        res = search.by_coordinate(y,x, radius=100, returns=5)
        for result in res:
            if(result["City"]==temp["city"]):
                # put new zipcode to the data
                market.loc[i, 'zip'] = result["Zipcode"]
                break
            else:
                market.loc[i, 'zip'] = result["Zipcode"]
                break
```

```python
zip_still_null =np.where(pd.isnull(market["zip"]))
print(len(*zip_still_null))
print(zip_still_null)
```

```python
# x or y values are null for those 27 datasets
for i in list(*zip_still_null):
    x = market.loc[i, "x"]
    print(x)
```
